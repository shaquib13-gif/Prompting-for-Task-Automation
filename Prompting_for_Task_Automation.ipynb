```python
# SkillCraft Technology - Task 03
# Prompting for Task Automation

print("===== Task Automation using Prompting =====\n")

# Example inputs
notes = [
    "Python is a programming language used for AI, web development, and automation.",
    "Machine Learning helps computers learn from data without explicit programming.",
    "GitHub is used to store and manage coding projects online."
]

# Function to summarize notes
def summarize(text):
    words = text.split()
    summary = " ".join(words[:8]) + "..."
    return summary

# Process each note
for i, note in enumerate(notes, start=1):
    print(f"Input Note {i}:")
    print(note)

    print("\nAutomated Summary:")
    print(summarize(note))

    print("\n" + "-"*60 + "\n")

print("Task Completed Successfully!")
```
